```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import preprocessing
from scipy.stats.mstats import winsorize
import warnings
warnings.filterwarnings('ignore')
```

```python
weather = pd.read_csv('C:\\Users\\Pc\\Desktop\\weatherHistory.csv')
weather.head()
```

**Hedef Değişken**

```python
weather['temp'] = weather['Temperature (C)'] - weather['Apparent Temperature (C)']
```

Doğrusal regresyon parametrelerini tahmin edelim.

**Model (Etkileşimsiz)**

```python
X = weather[['Humidity','Wind Speed (km/h)']]
Y = weather[['temp']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

Tahmin edilen modelde R kare değeri ve ayarlanmış R kare değeri 0.288'dir. Değer varyansın açıklanması için düşüktür.

**Model (Etkileşimli) Nem ve Rüzgar Hızı**

```python
weather['hwind'] = weather['Humidity'] * weather['Wind Speed (km/h)']

X1 = weather[['Humidity', 'Wind Speed (km/h)', 'hwind']]
Y1 =  weather[['temp']]
lrm = linear_model.LinearRegression()
lrm.fit(X1, Y1)
X1 = sm.add_constant(X1)
result1 = sm.OLS(Y1, X1).fit()
result1.summary()
```

Tahmin Modeli:

Sıcaklık=-0.0839-0.1775*nem-0.0905*rüzgarhızı+0.2971*hwind

Modelde tüm katsayılar istatistiksel olarak anlamlıdır. nem, rüzgar hızı ve etkileşimleri hissedilen sıcaklığı etkiler. R kare ve ayarlanmış R Kare 0.341 olmuştur. Bu modelimiz diğerine göre daha çok varyansı açıklıyor.

**Model(İlk Modele Görünürlük Dahil Edildi.)**

```python
X2 = weather[['Humidity','Wind Speed (km/h)', 'Visibility (km)']]
Y2 = weather[['temp']]

X2 = sm.add_constant(X2)

result2 = sm.OLS(Y2,X2).fit()
result2.summary()
```

R kare değeri 0.304 ilk modele göre yüksek ama ikinci modele göre düşük bir değerdir. İkinci model yani etkileşimli model hedef değişkenimizi daha iyi açıklıyor. İkinci model ve görünürlük modeli için de etkileşmli modelin R kare değeri daha fazladır. Etkileşimli model varyansı açıklamada bu modeller arasında en etkilisidir. 

İlk modelimiz için R-kare 0.288, ayarlanmış R-kare 0.288, AIC 3.409e+05 ve BIC 3.409e+05, ikinci modelimiz için ise R-kare 0.341, ayarlanmış R-kare 0.341, AIC 3.334e+05 ve BIC 3.334e+05, 3.modelimiz için R-kare 0.304, ayarlanmış R-kare 0.303, AIC 3.388e+05 ve BIC 3.388e+05'dir. Tüm metriklere göre, ikinci modelimiz diğerlerinden daha iyi görünüyor.

```python
data = pd.read_csv('C:\\Users\\Pc\\Desktop\\train.csv')
data.head(10)
```

```python
df=data.loc[:, ['Street', 'OverallQual', 'YearBuilt', 'ExterQual', 'SalePrice']]
df.head()
```

```python
df["Str"] = pd.get_dummies(df.Street, drop_first=True)
df["Ext"] = df.ExterQual.replace({'Gd':1, "TA":0, "Ex":2, "Fa":3})
df[['Str','Ext']].head()
```

Modeldeki YearBuilt ve OverallQual ve SalePrice değişkenlerini standartlaştıralım.

```python
from sklearn.preprocessing import scale
df['st_yearbuilt']=scale(df["YearBuilt"])
df['st_saleprice']=scale(df["SalePrice"])
df['st_ovqual']=scale(df["OverallQual"])
df.head()
```

**Model Tahmini**

```python
Y3=df['st_saleprice']
X3=df[['Str', 'st_ovqual', 'st_yearbuilt', 'Ext']]
lrm = linear_model.LinearRegression()
lrm.fit(X3, Y3)
X3 = sm.add_constant(X3)
result1 = sm.OLS(Y3, X3).fit()
result1.summary()
```

Street katsayısı hariç diğer katsayılar istatistiksel olarak anlamlıdır. hedef değişkenini yani konut fiyatlarını bina yaşı, bina genel malzemesi ve dış kaplama malzemesi etkilemektedir. R kare değeri 0.648, ayarlanmış R kare 0.647'dir Aslında yüksek bir değer ama Street özelliğini çıkarıp bakalım.

Street değişkenini modelimizden çıkaralım ve yeni model için tahmin katsayılarını bulalım.

```python
Y4 = df['st_saleprice']
X4 = df[['st_yearbuilt', 'Ext', 'st_ovqual' ]]
lrm = linear_model.LinearRegression()
lrm.fit(X4, Y4)
X4 = sm.add_constant(X4)
result1 = sm.OLS(Y4, X4).fit()
result1.summary()
```
